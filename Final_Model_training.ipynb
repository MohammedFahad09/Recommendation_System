```python
import pandas as pd
import ast
 
df=pd.read_json(r"C:\myfiles\CODING\Main_Project\Image-Based-Personalised-Book-Recommendatiion\Data\goodreads_books.json",lines=True,chunksize=200000)

result_chunks = []

for chunk in df:

    chunk["genre"] = chunk["popular_shelves"].apply(
        lambda shelves: [
            item["name"].lower().replace("-", " ")
            for item in shelves
            if item["name"].lower().replace("-", " ") in top_200_genres
        ]
    )


    # remove rows with no genre
    chunk = chunk[chunk["genre"].apply(len) > 0]
    important_cols = [
    "book_id",
    "title",
    "title_without_series",
    'description',
    'authors',
    "average_rating",
    "ratings_count",
    "text_reviews_count",
    "image_url",
     "genre"
    ]
    # [
    #     'book_id', 'title', 'title_without_series',
    #     'description', 'authors', 'popular_shelves',
    #     'average_rating', 'ratings_count', 'image_url'
    # ]
    chunk[important_cols]

    result_chunks.append(chunk[important_cols])
    

final_df = pd.concat(result_chunks, ignore_index=True)
```

```python
import ast

def process_genre(x):
    if isinstance(x, str):
        try:
            x = ast.literal_eval(x)
        except:
            return ""
    
    if isinstance(x, list):
        return " ".join(x)
    
    return ""

final_df["genre"] = final_df["genre"].apply(process_genre)
```

```python
import pandas as pd
df=pd.read_csv(r'C:\myfiles\CODING\Main_Project\Image-Based-Personalised-Book-Recommendatiion\Noteboooks\final_books2.csv')
df.head()
```

```python
import re

def clean_title(x):
    x = str(x).lower()
    x = re.sub(r"\(.*?\)", "", x)   # remove series
    x = re.sub(r"[^a-z0-9 ]", "", x)
    return x.strip()

df["clean_title"] = df["title_without_series"].apply(clean_title)
```

```python
def clean_text(x):
    x = str(x).lower()
    x = re.sub(r"[^a-z0-9 ]", " ", x)
    return x

df["clean_description"] = df["description"].apply(clean_text)
```

```python
df = df.drop(columns=["authors"])
```

```python
df["features"] = (
    df["clean_title"] + " " +
    df["clean_description"] + " " +
    df["genre"]
)
```

```python
df = df.drop(columns=["description",'title_without_series'])
```

```python
import pickle
df["features"] = (
    df["clean_title"].astype(str) + " " +
    df["clean_description"].astype(str) + " " +
    df["genre"].astype(str)
)

# ---------------- LOAD TFIDF ---------------- #

with open("model/tfidf.pkl", "rb") as f:
    tfidf = pickle.load(f)

# ---------------- CREATE TFIDF MATRIX ---------------- #

tfidf_matrix = tfidf.transform(df["features"])

print("✅ TF-IDF matrix created")

# ---------------- CREATE MAPPINGS ---------------- #

titles = df["clean_title"].tolist()

title_to_index = {title: idx for idx, title in enumerate(titles)}

print("✅ Mapping created")

# ---------------- SAVE Tfidf ---------------- #

pickle.dump(tfidf, open('tfidf2.pkl', 'wb'))


# ---------------- SAVE MATRIX ---------------- #

from scipy.sparse import save_npz
save_npz("tfidf2_matrix.npz", tfidf_matrix)

# # ---------------- SAVE METADATA ---------------- #

with open("metadata2.pkl", "wb") as f:
    pickle.dump({
        "df": df,
        "titles": titles,
        "title_to_index": title_to_index
    }, f)

print("✅ metadata saved")
```
